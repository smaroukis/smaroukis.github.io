```python
# env:figs
# import plotly # 2.4.1
import plotly.offline
import plotly.plotly as py
import plotly.graph_objs as go
import numpy as np
```

```python

1/(2*60.)
```

```python
x=np.arange(0.5,30.1,0.1)
y1=np.sqrt(1+2*np.exp(-2*120*np.pi*1/60/1*1/x))
y2=np.sqrt(1+2*np.exp(-2*120*np.pi*1/60/2*1/x))
y3=np.sqrt(1+2*np.exp(-2*120*np.pi*1/60/3*1/x))
y4=np.sqrt(1+2*np.exp(-2*120*np.pi*1/60/4*1/x))
```

```python
plotly.offline.init_notebook_mode(connected=True) # includes js files
plotly.offline.iplot({
        'data': [go.Scatter(
            x=x,
            y=y1,
            mode='markers+lines',
            name='t=1/60s'),
                go.Scatter(
            x=x,
            y=y2,
            mode='markers+lines',
            name='t=1/120s'),
                go.Scatter(
            x=x,
            y=y4,
            mode='markers+lines',
            name='t=1/240s')
        ],
        'layout': {
            'title': "Curve of Asymmetry Factors",
            'yaxis': {'type': 'log'},
            'xaxis': {'type': 'log',
                     'title': 'X/R'}
        }
    })
```
